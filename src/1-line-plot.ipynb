### Exercise 1: Line Plot Customization

```python
# TASK: Create a line plot with x values ranging from 0 to 10 and y values as the square of x.
# Customize the plot by adding a title, labels for both axes, and a grid.
import matplotlib.pyplot as plt
import numpy as np

try:
    # Generate x values from 0 to 10
    x = np.linspace(0, 10, 100)
    
    # Ensure x is correctly generated
    if x is None or len(x) == 0:
        raise ValueError("x values are not correctly generated.")
    
    # Calculate y values as the square of x
    y = x ** 2

    # Ensure y is correctly generated
    if y is None or len(y) == 0:
        raise ValueError("y values are not correctly generated.")

    # Create the line plot
    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label='y = x²', color='blue')

    # Customize the plot
    plt.title('Line Plot of y = x²', fontsize=16)
    plt.xlabel('x values', fontsize=14)
    plt.ylabel('y values (y = x²)', fontsize=14)
    plt.grid(True)
    plt.legend()

    # Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
