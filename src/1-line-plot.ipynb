### Exercise 1: Line Plot Customization

```python
import matplotlib.pyplot as plt

try:
    # Step 1: Define x and y values
    x = [0, 1, 2, 3, 4, 5]
    y = [0, 1, 4, 9, 16, 25]

    # Step 2: Validate data
    if len(x) == 0 or len(y) == 0:
        raise ValueError("x or y values cannot be empty.")
    if len(x) != len(y):
        raise ValueError("x and y lists must have the same length.")

    # Step 3: Create the plot
    plt.figure(figsize=(8, 6))
    plt.plot(x, y, marker='o', linestyle='-', color='b', label='y = x²')

    # Step 4: Customize the plot
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Basic Line Plot')
    plt.grid(True)
    plt.legend()

    # Step 5: Show the plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
