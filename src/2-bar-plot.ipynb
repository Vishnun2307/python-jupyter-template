### Exercise 2: Bar Plot with Colors

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category A','Category B','Category C','Category D']
values = [4, 7, 3, 6]

try:
    # Check if categories and values are valid
    if not categories or not values:
        raise ValueError("Categories or values cannot be empty.")

    if len(categories) != len(values):
        raise ValueError("Categories and values must have the same length.")

    # Create bar plot
    plt.figure(figsize=(8, 5))
    plt.bar(categories, values, color=['blue', 'orange', 'green', 'red'])

    # Add title and labels
    plt.title('Bar Plot Example', fontsize=14)
    plt.xlabel('Categories', fontsize=12)
    plt.ylabel('Values', fontsize=12)
    
    # Show plot
    plt.show()

except Exception as e:
    print(f"An error occurred: {e}")
```
