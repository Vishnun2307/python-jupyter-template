### Exercise 6: Histogram Customization

```python
# TASK: Create a histogram for the following data: data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5].
# Customize the histogram with a title, labels for the x-axis, and a specific color for the bars.
import matplotlib.pyplot as plt  

# Data  
data = [1, 2, 2, 3, 3, 3, 4, 4, 4, 4, 5]  

try:
    # Validate data
    if not data:
        raise ValueError("Data cannot be empty.")

    if not all(isinstance(i, (int, float)) for i in data):
        raise ValueError("All elements in data must be numeric.")

    # Create a histogram  
    plt.figure(figsize=(8, 6))  
    plt.hist(data, bins=range(1, 7), color='skyblue', edgecolor='black', alpha=0.7)  

    # Customizing the histogram  
    plt.title('Histogram of Data', fontsize=14)  
    plt.xlabel('Value', fontsize=12)  
    plt.ylabel('Frequency', fontsize=12)  

    # Show the plot  
    plt.grid(axis='y', alpha=0.75)  # Optional: add grid for better readability  
    plt.tight_layout()  
    plt.show()  

except Exception as e:
    print(f"An error occurred: {e}")
```
